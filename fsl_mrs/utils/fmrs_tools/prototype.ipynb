```python
%load_ext autoreload
%autoreload 2

from fsl_mrs.dynamic.dyn_results import load_dyn_result, ResultLoadError
import pandas as pd
from pathlib import Path
from dataclasses import dataclass
import numpy as np

res_dir = Path('/Users/wclarke/Documents/Python/fmrs_demo/first_level_results/sub0/stim')

# Either load full results object (if data availible) or just the key dataframes
try:
    obj = load_dyn_result(res_dir)
    value_df = obj.dataframe_free
    cov_df = obj.cov_free
except ResultLoadError:
    value_df = pd.read_csv(res_dir / 'dyn_results.csv', index_col=0)
    cov_df = pd.read_csv(res_dir / 'dyn_cov.csv', index_col=0)
cov_df
```

```python
peaks_comb = [['PCh','GPC'],
              ['Cr','PCr'],
              ['NAA', 'NAAG'],
              ['Glu', 'Gln', 'GABA']]

@dataclass
class contrast:
    name: str
    betas: list
    scale: list

contrasts = [
    contrast('mean', ['beta0', 'beta1'], [0.5, 0.5]),
    contrast('0-1', ['beta0', 'beta1'], [1, -1]),
    contrast('1-0', ['beta0', 'beta1'], [-1, 1])]
    

```

```python
# Three cases to identify groupings:
# 1. metabolites to combine - this is the sum of two or more metabolites. Applied to all betas in turn
# 2. Linear combination of betas - apply to all metabolites in turn
# 3. The combination of 1 & 2 which should be performed in a single step to capture all covariances

all_params = value_df.columns.to_list()

import re
metab_re = re.compile(r'^conc_(.*?)_',)

metabolites = np.array([metab_re.match(param)[1] for param in all_params if metab_re.match(param)])
metabolites = np.unique(metabolites)

metab_beta_dict = {}
for metab in metabolites:
    beta_re = re.compile(rf'^conc_{metab}_(.*)$',)
    betas = [beta_re.match(param)[1] for param in all_params if beta_re.match(param)]
    metab_beta_dict[metab] = betas
metab_beta_dict


# value_df.filter(regex='Asp')

```

```python
def comb_variance(cov, scale):
    """Calculate variance of linear combinations of scalar values

    :param cov: parameter covariance matrix
    :type cov: numpy.array
    :param scale: scale vector (n x 1) vector of parameter scalings
    :type scale: numpy.array
    :return: combined variance
    :rtype: float
    """

    if isinstance(cov, pd.DataFrame):
        return np.asarray(scale) @ cov.to_numpy() @ np.asarray(scale)
    else:
        return np.asarray(scale) @ cov @ np.asarray(scale)

def comb_value(df, parameters, name, scale=None):
    if scale is None:
        scale = np.ones(len(parameters))
    
    return pd.DataFrame(
        df[parameters].multiply(scale).sum(axis=1),
        columns=[name])
    
```

```python
class MismatchedBetasError():
    pass
new_params = []
new_var = {}

# 1. metabolites to combine - this is the sum of two or more metabolites. Applied to all betas in turn
for metabs in peaks_comb:
    for met in metabs:
        if not metab_beta_dict[met] == metab_beta_dict[metabs[0]]:
            raise MismatchedBetasError(f'{met} has betas {metab_beta_dict[met]}, but {metabs[0]} has different betas: {metab_beta_dict[metabs[0]]}.')
    
    new_met_name = '+'.join(metabs)
    # print(new_met_name)
    # loop over betas
    for beta in metab_beta_dict[metabs[0]]:
        parameters = [f'conc_{met}_{beta}' for met in metabs]
        
        new_param_name = f'conc_{new_met_name}_{beta}'
        # Parameter value (sum)
        new_params.append(comb_value(value_df, parameters, new_param_name))

        # Parameter variance
        n_met = len(parameters)
        new_var[new_param_name] = comb_variance(cov_df.loc[parameters,parameters], np.ones(n_met))
```

```python
# 2. Linear combination of betas - apply to all metabolites in turn

for con in contrasts:
    # Loop over metabolites
    for met in metabolites:
        if not all([x in metab_beta_dict[met] for x in con.betas]):
            raise MismatchedBetasError(f'Contrast {con.name} for {met} requires betas {con.betas}, but {met} has different betas: {metab_beta_dict[met]}.')  
        parameters = [f'conc_{met}_{beta}' for beta in con.betas]

        new_param_name = f'conc_{met}_{con.name}'
        # Parameter value (sum)
        new_params.append(comb_value(value_df, parameters, new_param_name, scale=con.scale))

        # Parameter variance
        new_var[new_param_name] = comb_variance(cov_df.loc[parameters,parameters], con.scale)
```

```python
# 3. The combination of 1 & 2 which should be performed in a single step to capture all covariances
for con in contrasts:
    for metabs in peaks_comb:
        parameters = [f'conc_{met}_{beta}' for beta in con.betas for met in metabs]
        new_met_name = '+'.join(metabs)
        new_param_name = f'conc_{new_met_name}_{con.name}'
        # print(new_param_name)
        # print(parameters)
        scale = np.stack([np.array(con.scale) for _ in range(len(metabs))]).T.ravel()
        # print(scale)
        # Parameter value (sum)
        new_params.append(comb_value(value_df, parameters, new_param_name, scale=scale))

        # print(value_df[parameters])
        # print(new_params[-1])
        # Parameter variance
        new_var[new_param_name] = comb_variance(cov_df.loc[parameters,parameters], scale)
```

```python
# Now add to existing data frames
new_value_df = pd.concat([value_df, ] + new_params, axis=1)
new_cov_df = cov_df.copy()
for var in new_var:
    new_cov_df.loc[var, var] = new_var[var]
new_cov_df
```

```python
new_value_df.mean()
pd.Series(np.diag(new_cov_df), index=[new_cov_df.index])

mean_free = new_value_df.mean()
std_free = pd.Series(np.diag(new_cov_df), index=new_cov_df.index)
pd.concat((mean_free, std_free), axis=1, keys=['mean', 'sd'])
```
